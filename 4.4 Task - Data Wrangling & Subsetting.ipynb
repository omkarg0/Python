```python
# After creating an index column in the previous step, we can create a new header for the dataframe. 
# This will be done in three steps: 
# 1) Create a new header; 
# 2) Remove the first row in the dataframe; and 
# 3) Add the new header. This all sounds a little abstract, so let’s walk through the process together.
new_header = df_dep_t.iloc[0]
```

```python
new_header # see the contents of your first row
```

```python
df_dep_t_new = df_dep_t[1:] # Now that you’ve made a copy of your first row, you need to delete it. Otherwise, when you add the first row back in as your header, you’ll have two of them: one as a header and one as a row. That’s not what you want! As such, let’s create a new dataframe that only copies over rows beyond the first row.
```

```python
df_dep_t_new # The resulting df_dep_t_new dataframe will look exactly like the old dataframe only without the first row.
```

```python
df_dep_t_new.columns = new_header # set your new_header variable as the new header for your df_dep_t_new dataframe
```

```python
df_dep_t_new # Proper column headers!
```

```python
df_snacks =  df_prods[df_prods['department_id']==19] # save the results above into a new dataframe called df_snacks
```

```python
df_snacks.head() # check out the dataframe
```

```python
df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19] # A second way of achieving the above same results which is a bit simpler. This involves use of the loc function. The loc function is similar to the iloc function you used previously, but while iloc looks for an integer position, loc looks for a column name. In your case, you want this function to look for instances where the value of "department_id" is 19 within the df_prods database
```

```python
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])] # Another way to achive the same results. In this second version, rather than using an equals sign, you’re telling the loc function to look into a list: isin([19]). This is handy because you could use more than one value here if you wanted to check multiple "department_ids" at the same time (isin([17,18,19])).
```
